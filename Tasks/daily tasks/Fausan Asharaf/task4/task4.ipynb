```python
import torch
import torchvision
import torchvision.transforms as transforms
from PIL import Image
import matplotlib.pyplot as plt
```

```python
flowerIm = Image.open('./flower/flower5.jpg')
plt.imshow(flowerIm)
```

```python
transform = transforms.Compose([transforms.CenterCrop(100)])
newIm = transform(flowerIm)
plt.imshow(newIm)
```

```python
transform = transforms.Compose([transforms.ColorJitter(brightness=100.0, contrast=20.5, saturation=0)])
newIm = transform(flowerIm)
plt.imshow(newIm)
```

```python
transform = transforms.Grayscale(1)
newIm = transform(flowerIm)
plt.imshow(newIm)
print(newIm.size)
```

```python
flowerIm = [Image.open('./flower/flower1.jpg'),Image.open('./flower/flower2.jpg'),
            Image.open('./flower/flower3.jpg'), Image.open('./flower/flower4.jpg'), Image.open('./flower/flower5.jpg')]
fig = plt.figure(figsize=(12, 48))
for i in range(5):
    ax = fig.add_subplot(1, 5, i+1, xticks=[], yticks=[])
    plt.imshow(flowerIm[i])
    ax.set_title('Flower {0}'.format(i+1))
```

```python
transform = transforms.RandomApply([transforms.RandomCrop(size=500), transforms.RandomVerticalFlip(), transforms.ColorJitter(saturation=11.1, hue=0.35)])

fig = plt.figure(figsize=(12, 48))
for i in range(5):
    ax = fig.add_subplot(1, 5, i+1, xticks=[], yticks=[])
    plt.imshow(transform(flowerIm[i]))
    ax.set_title('Flower {0}'.format(i+1))
```

```python
transform = transforms.Pad(padding=20, fill=(0, 255, 0))

fig = plt.figure(figsize=(12, 48))
for i in range(5):
    ax = fig.add_subplot(1, 5, i+1, xticks=[], yticks=[])
    plt.imshow(transform(flowerIm[i]))
    ax.set_title('Flower {0}'.format(i+1))
```

```python
transform = transforms.Resize(size = (500, 500))

fig = plt.figure(figsize=(12, 48))
for i in range(5):
    ax = fig.add_subplot(1, 5, i+1, xticks=[], yticks=[])
    plt.imshow(transform(flowerIm[i]))
    newIm = transform(flowerIm[i])
    w, h = newIm.size
    ax.set_title('Flower {0} - ({1}, {2})'.format(i+1, w, h))
```

```python
import numpy as np
transform = transforms.Compose([transforms.ToTensor(), transforms.RandomErasing(p=0.9)])

def imshow(img, one_channel=False):
    if one_channel:
        img = img.mean(dim=0)
    img = img / 2 + 0.5     # unnormalize
    npimg = img.numpy()
    if one_channel:
        plt.imshow(npimg, cmap="Greys")
    else:
        plt.imshow(np.transpose(npimg, (1, 2, 0)))

fig = plt.figure(figsize=(12, 48))
for i in range(5):
    ax = fig.add_subplot(1, 5, i+1, xticks=[], yticks=[])
    imshow(transform(flowerIm[i]))
    ax.set_title('Flower {0}'.format(i+1))
```

```python
transform = transforms.Compose([transforms.ToTensor(), transforms.RandomErasing(p=0.9), transforms.ToPILImage()])


fig = plt.figure(figsize=(12, 48))
for i in range(5):
    ax = fig.add_subplot(1, 5, i+1, xticks=[], yticks=[])
    plt.imshow(transform(flowerIm[i]))
    ax.set_title('Flower {0}'.format(i+1))
```

```python
transform = transforms.Compose([transforms.ToTensor(), transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5)),
                                transforms.ToPILImage()])


fig = plt.figure(figsize=(12, 48))
for i in range(5):
    ax = fig.add_subplot(1, 5, i+1, xticks=[], yticks=[])
    plt.imshow(transform(flowerIm[i]))
    ax.set_title('Flower {0}'.format(i+1))
```
