```
import torch
import numpy as np
```

```
q=np.random.rand(5,3)
print(q)
w=np.random.rand(3,4)
print(w)
```

```
e=torch.from_numpy(q)
e.mul_(0);e.add_(2)
print(e)
print(q)
```

```
r=torch.tensor(w)
r.mul_(0);r.add_(2)
print(r)
print(w)
```

```
y=np.matmul(q,r)
y
```

```
z=torch.matmul(e,r)
z
```
